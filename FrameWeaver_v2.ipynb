```python
import subprocess

def get_fps(video_path):
    cmd = ["ffprobe", 
           "-v", "error", 
           "-select_streams", "v:0", 
           "-show_entries", "stream=r_frame_rate", 
           "-of", "default=noprint_wrappers=1:nokey=1", 
           video_path]

    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    out, err = process.communicate()

    if err:
        raise Exception(err.decode('utf-8'))

    fps_string = out.decode('utf-8').strip()
    if "/" in fps_string:
        num, den = fps_string.split("/")
        return float(num) / float(den)
    else:
        return float(fps_string)
```

```python
video_path = fName1+'.mov'
fps = get_fps(video_path)
print("FPS:", fps)
```

```python
video_path = fName2+'.mov'
fps = get_fps(video_path)
print("FPS:", fps)
```

```python
def run_ffmpeg(input_file, output_file, **kwargs):
    """Runs FFmpeg with the given input file, output file, and arguments."""

    # Build the FFmpeg command
    command = ['ffmpeg', '-i', input_file]

    # Add any additional arguments
    for key, value in kwargs.items():
        if value is not None:
            command.extend(['-' + key, str(value)])

    command.append(output_file)

    # Execute the command
    subprocess.run(command)
```

```python
run_ffmpeg(fName1+'.mov', fName1+'.mp4', vcodec='libx264')
```

```python
import cv2
import numpy as np
import os

# set video file path of input video with name and extension

imgDirName = 'imgFrames'+fName1+fName2


if not os.path.exists(imgDirName):
    os.makedirs(imgDirName)

#---------------------------------------------------------    

vid = cv2.VideoCapture(fName1+'.mp4')

#for frame identity
index = 0
while(True):
    # Extract images
    ret, frame = vid.read()
    # end of frames
    if not ret: 
        break
    # Saves images
    name = imgDirName+'/frame' + str(100000+index) + '.jpg'
    print ('Creating ' + name)
    cv2.imwrite(name, frame)

    # next frame
    index += 2
    
#---------------------------------------------------------    
    
vid = cv2.VideoCapture(fName2+'.mp4')

index = 1
while(True):
    # Extract images
    ret, frame = vid.read()
    # end of frames
    if not ret: 
        break
    # Saves images
    name = imgDirName+'/frame' + str(100000+index) + '.jpg'
    print ('Creating ' + name)
    cv2.imwrite(name, frame)

    # next frame
    index += 2
```

```python
import subprocess

output_video = imgDirName+'.mp4'

command = [
    'ffmpeg',
    '-framerate', '60',  # Set your desired frame rate
    '-i', f'{imgDirName}/frame1%05d.jpg',  # Pattern for frame names
    '-c:v', 'libx264',  # Use H.264 codec
    '-pix_fmt', 'yuv420p',  # Set pixel format
    output_video
]

subprocess.run(command, check=True)
```
