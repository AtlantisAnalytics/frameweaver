```python
import cv2
import numpy as np
import os

# set video file path of input video with name and extension


if not os.path.exists('imgFrames'):
    os.makedirs('imgFrames')

#---------------------------------------------------------    

vid = cv2.VideoCapture('A.MP4')

#for frame identity
index = 0
while(True):
    # Extract images
    ret, frame = vid.read()
    # end of frames
    if not ret: 
        break
    # Saves images
    name = 'imgFrames/frame' + str(100000+index) + '.jpg'
    print ('Creating ' + name)
    cv2.imwrite(name, frame)

    # next frame
    index += 2
    
#---------------------------------------------------------    
    
vid = cv2.VideoCapture('B.MP4')

index = 1
while(True):
    # Extract images
    ret, frame = vid.read()
    # end of frames
    if not ret: 
        break
    # Saves images
    name = 'imgFrames/frame' + str(100000+index) + '.jpg'
    print ('Creating ' + name)
    cv2.imwrite(name, frame)

    # next frame
    index += 2
```
